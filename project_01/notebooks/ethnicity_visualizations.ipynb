```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
race_dict = {'Black': ['AfricanAmerican', 'Ugandan'],
             'Asian': ['Indian', 'Chinese'],
             'Hispanic/Latino': ['Mexican']
            }
```

```python
decade_dict = {0:2000,
               1:2010,
               3:1930,
               4:1940,
               5:1950,
               6:1960,
               7:1970,
               8:1980,
               9:1990
              }
```

```python
df = pd.read_csv('actorsWithEthnicity.csv')
df.tail(100)
```

```python
def create_decade(year):
    '''
    year: int
    '''
    v = str(year)[2]
    return decade_dict[int(v)]
```

```python
df['Decade'] = df['Year'].apply(create_decade)
```

```python
def create_race(ethnicities):
    '''
    ethnicities: list of strings
    '''
    ethnicity_string = ''.join(str(elem) for elem in ethnicities)
    for k in race_dict.keys():
        for e in race_dict[k]:
            if e in ethnicity_string:
                return k
    return 'White'
```

```python
df['Race'] = df['Ethnicity'].apply(create_race)
```

```python
# Ethnicities across all nominated (total years)
plt.bar(df['Race'].value_counts().index, df['Race'].value_counts())
plt.title('Race of all Nominated Actors 1934-2019')
```
