```python
# Imports

import numpy as np
import pandas as pd
import seaborn as sns
import os

%matplotlib inline
```

```python
# Sample dataset of 2016 commercials

datadir = 'data'
path = os.path.join(datadir, str(2016))
df = pd.read_csv(os.path.join(path, str(2016)+'_features.csv'), index_col=0)
```

```python
# Generate sample labels

test_df = df.copy()
test_df['SB'] = test_df['0']
test_df = test_df.drop(['0'], axis=1)
new_sbs = list()
for i in range(len(test_df)):
    if i < 55:
        new_sbs.append('SB')
    else:
        new_sbs.append('Non-SB')
test_df['Super Bowl?'] = new_sbs
test_df = test_df.drop(['1', 'SB'], axis=1)
```

```python
test_df.head()
```

```python
# Scatterplot of feature mean to standard deviation
ax = sns.scatterplot(test_df['2'], test_df['3'], hue=test_df['Super Bowl?'])
ax.set_title('Spectral Centroid Mean vs. Spectral Centroid Standard Deviation')
ax.set_xlabel('Spectral Centroid Mean')
y = ax.set_ylabel('Spectral Centroid Std. Dev.')
```

```python
# Scatterplot of feature mean to another mean
ax = sns.scatterplot(test_df['2'], test_df['4'], hue=test_df['Super Bowl?'])
ax.set_title('Spectral Centroid Mean vs. Spectral Bandwidth Mean')
ax.set_xlabel('Spectral Centroid Mean')
y = ax.set_ylabel('Spectral Bandwidth Mean')
```

```python
# Scatterplot of feature standard deviation to another standard deviation
ax = sns.scatterplot(test_df['3'], test_df['5'], hue=test_df['Super Bowl?'])
ax.set_title('Spectral Centroid Standard Deviation vs. Spectral Bandwidth Standard Deviation')
ax.set_xlabel('Spectral Centroid Std. Dev')
y = ax.set_ylabel('Spectral Bandwidth Std. Dev.')
```

```python
# Correlation heatmap of all features
ax = sns.heatmap(test_df.drop('Super Bowl?', axis=1).corr())
ax.set_title('All Audio Features')
```

```python
# Only using time-wise means
ax = sns.heatmap(test_df[[str(i) for i in range(2, 102, 2)]].corr())
```
